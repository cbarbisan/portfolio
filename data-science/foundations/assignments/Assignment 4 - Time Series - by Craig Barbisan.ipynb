```python
import pandas as pd
import pandas_datareader.data as web
import datetime
import requests_cache

# Setup a local cache so it doesn't have to download the data every time.
expire_after = datetime.timedelta(days=3)
session = requests_cache.CachedSession(cache_name='cache', backend='sqlite', expire_after=expire_after)

# Set common parameters for the stock quote downloads
source = 'yahoo'
start = datetime.datetime(2012, 7, 31) 
end = datetime.datetime(2017, 6, 30)

# Download quote data for each of the 4 stocks
aapl = web.DataReader('AAPL', source, start, end, session=session)
bbry = web.DataReader('BBRY', source, start, end, session=session)
lulu = web.DataReader('LULU', source, start, end, session=session)
amzn = web.DataReader('AMZN', source, start, end, session=session)
amzn
```

```python
# Build a DataFrame of resampled stock quote data.
# The time series index will transform from daily, to last business day of each month.
# Each column in the DataFrame will contain the month-end adjusted close price of a particular stock.
monthly_adj_close = pd.DataFrame()
monthly_adj_close['AAPL'] = aapl['Adj Close'].resample('BM').last()
monthly_adj_close['AMZN'] = amzn['Adj Close'].resample('BM').last()
monthly_adj_close['BBRY'] = bbry['Adj Close'].resample('BM').last()
monthly_adj_close['LULU'] = lulu['Adj Close'].resample('BM').last()
monthly_adj_close
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from pandas.plotting import autocorrelation_plot

# Plot the autocorrelation of the monthly adjusted close price of each stock.
# We will plot them on the same figure.
plt.figure()
autocorrelation_plot(monthly_adj_close.AAPL)
autocorrelation_plot(monthly_adj_close.AMZN)
autocorrelation_plot(monthly_adj_close.BBRY)
autocorrelation_plot(monthly_adj_close.LULU)

# Add an appropriate title
plt.title("Autocorrelation for Monthly Adjusted Close Price")

# Setup the legend
blue_patch = mpatches.Patch(label="AAPL")
orange_patch = mpatches.Patch(color="orange", label="AMZN")
green_patch = mpatches.Patch(color="green", label="BBRY")
red_patch = mpatches.Patch(color="red", label="LULU")
plt.legend(handles=[blue_patch, orange_patch, green_patch, red_patch])

# Label the axes
plt.xlabel("Lags")
plt.ylabel("ACF")

# Show the plot
plt.show()
```

The above visualization tells us that the month-end adjusted closing price of the stocks are auto-correlated, since one or more of the auto-correlation values are significantly non-zero for each of the stocks listed. By "significantly non-zero", we mean outside of the confidence bands (auto-correlations within the confidence bands are insignificant). The time series is non-random.

```python
# Calculate the monthly returns for each stock as a percentage.
# Shifting the monthly adjusted close prices in the denominator results
# in a row of NaN values for the first month, so that row is dropped.
monthly_returns = (monthly_adj_close/monthly_adj_close.shift(1) - 1).dropna()
monthly_returns
```

```python
# Plot the autocorrelation of the monthly returns for each stock.
# We will plot them on the same figure.

plt.figure()
autocorrelation_plot(monthly_returns.AAPL)
autocorrelation_plot(monthly_returns.AMZN)
autocorrelation_plot(monthly_returns.BBRY)
autocorrelation_plot(monthly_returns.LULU)

# Add an appropriate title
plt.title("Autocorrelation for Monthly Returns")

# Setup the legend
blue_patch = mpatches.Patch(label="AAPL")
orange_patch = mpatches.Patch(color="orange", label="AMZN")
green_patch = mpatches.Patch(color="green", label="BBRY")
red_patch = mpatches.Patch(color="red", label="LULU")
plt.legend(handles=[blue_patch, orange_patch, green_patch, red_patch])

# Label the axes
plt.xlabel("Lags")
plt.ylabel("ACF")

# Show the plot
plt.show()
```

The above plot shows that monthly returns are not-autocorrelated since none of the returns are outside of the confidence bands. The time series is random for each.
