```python
# Exercise #1: Create a plot where y = x**2 for x over the interval 1 to 5, properly labelled
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

# Start here
xaxis = [0,5]
yaxis = [0,25]
plt.axis(xaxis + yaxis)
plt.title("Exercise #1 - Plotting y = $x^2$")
blue_patch = mpatches.Patch(color="blue", label="$y = x^2$")
plt.legend(handles=[blue_patch])
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
x = np.fromiter(range(1,6), dtype="int")
y = x ** 2
plt.plot(x,y,'bo-')
plt.show()
```

```python
# Exercise #2: Create a histogram where the mean = 0, std. dev. = 1, n = 300, and there are sqrt(n) bins

import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline 

# Start here
mean = 0
stddev = 1
outcomes = 300
bin_ct = int(np.sqrt(outcomes))
data = mean + stddev * np.random.randn(outcomes)
n, bins, patches = plt.hist(data,bin_ct,normed=1)
plt.title("Exercise #2 - Histogram with $\mu = 0\ and\ \sigma = 1$")
plt.xlabel("Outcomes")
plt.ylabel("Probability")
plt.show()

# The chart will be a little different each time you run it
```

```python
# Exercise #3: (1) Create a line plot of your choosing with an appropriate legend which displays
# the formula of the curve depicted. (2) The formula should be formatted using TeX syntax
# to create a professional presentation. (3) Finally save your plot as a PDF file. 

import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline 

# Start here

x = np.linspace(-1, 1, 10001)
y = np.fabs(x) + np.sqrt(1-(x**2))
plt.title("Exercise #3 - Plotting a Curve")
yellow_patch = mpatches.Patch(color="yellow", label="$y = |x| + \sqrt{1 - x^2}$")
plt.legend(handles=[yellow_patch])
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(x, y, color='yellow')
plt.savefig('curveplot.pdf')
plt.show()
```
