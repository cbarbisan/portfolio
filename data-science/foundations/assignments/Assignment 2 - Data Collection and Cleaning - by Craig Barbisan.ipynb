The data is starting to look pretty good at this point. We've indexed the data on PassengerID and dropped the Cabin and Fare columns. Count the number of surviving and deceased passengers by gender

```python
df2.groupby(by=['Survived','Sex']).Sex.count()
```

This showed that despite the fact that there was twice as many males on board, almost 6x as many males died as females. And there were twice as many female survivors as male survivors. So it's pretty clear which gender was most likely to survive. But we have other data too...

Pclass is provided as an indicator of socio-economic status. It contains 3 distinct integer values, 1, 2 and 3 to indicate Upper, Middle and Lower class respectively. It may be useful to indicate this status using a more descriptive value.

```python
classType = {1:'Upper', 2:'Middle', 3:'Lower'}

def getSES(typecd):
    return classType[typecd]

df2['SES'] = df2.apply(lambda row: getSES(row['Pclass']), axis=1)
df2
```

We can look for survivors based on SES.

```python
df2.groupby(by=['Survived','SES']).SES.count()
```

This clearly shows that a disproportionate number of lower class passengers died relative to the other socio-economic groups. We could have run the same grouped count using Pclass, but SES gave us more descriptive output.

Now we can combine dimensions of our analysis for additional insight:

```python
df2.groupby(by=['Survived','Sex','SES']).SES.count()
```

Looking at upper class females, 91 survived and 3 died.<br/>
Looking at middle class females,70 survived and 6 died.<br/>
Looking at lower class females, 72 survived and <b><u>72</u> died!</b> Socio-economic status clearly impacted passenger survival on the Titanic.

The Embarked attribute represents the port of embarkation. It contains only 3 valid values, C, Q and S:<br/>
C = Cherbourg, Q = Queenstown, S = Southampton

Similar to what we did with the Pclass attribute, we can assign more descriptive values that represent the same data, which will improve the quality of our output. We know already from above that the Embarked attribute isn't full populated, but it is only missing 2 values in the whole set, which should not impact any results significantly. Our grouped count calculation ignores null values anyway:

```python
port = {'C':'Cherbourg', 'Q':'Queenstown', 'S':'Southampton'}

def getPort(portcd):
    if not pd.isnull(portcd):
        return port[portcd]

df2['Port'] = df2.apply(lambda row: getPort(row['Embarked']), axis=1)
df2
```

Now we can look for a relationship between port of embarkment and survival:

```python
df2.groupby(by=['Survived','Port']).Port.count()
```

```python
[(93/(93+75)*100),(30/(30+47)*100),(217/(217+427))*100]
```

Cherbourg survival: 55%, Queenstown survival: 38%, Southampton survival: 33%

A higher percentage of passengers from Southampton survived than Queenstown or Cherbourg. Based on what we've already learned, this may tell us that Southampton is more affluent than Queenstown and Cherbourg. Or it may tell us that more women embarked from Southampton than from the other 2 ports. We can test those assumptions:

```python
womenSurvivorsByPort = df2[(df2['Survived'] == 1) & (df2['Sex']=='female')].groupby(by=['Port']).Port.count()
womenSurvivorsByPort
```

These results validate the assumption that more women embarked from Southampton.

Now to test if Southampton is the more affluent port:

Another variable we can examine is age. Were children more likely to die since they have less strength than adults or because they haven't learned to swim yet? Or were children more likely to survive because room was made for them in the life boats? Were the elderly cared for in favour of younger adults who were more capable of looking after themselves? We can look for answers in the data to determine if a passenger's age was a factor in their survival or demise. But as we learned above, we are missing about 20% of our age values. One approach could be to just discard those records and not consider those passengers in the analysis. But the more data
we have, the more proof we have to support our conclusions.

Even if an age value is missing, there may be clues in our passenger records which give us an indication of their relative age. Some of the clues I have found include:
1. A female who uses the "Mrs." honorific is either married, divorced or a widow. In those cases, she is likely an adult.
2. A male who uses the "Master." honorific is a child. This honorific is typically only used for male children.
3. A male who uses the "Mr." honorific is an adult. If they were a child, "Master." would be used.
4. A passenger who is travelling alone is most likely not a child, thus an adult.

We will engineer an "AgeCategory" attribute, whose valid values are "Child", "Adult" or None (for unknown) and use it as a dimension in our analysis.

For starters, Age is typically the best indicator of whether or not a person is a child or an adult. But what does "adult" mean? Biological adulthood starts in the early teens or before. Legal adulthood in England in 1912 was 21 years of age. For the purpose of this analysis, we will split the difference and consider 16 to be the age of adulthood and anyone younger would be considered a child.

By default, set all AgeCategory values to None. We can update the attribute as we learn more about the passengers.

```python
df2['AgeCategory'] = None
```

Now apply the age test to determine AgeCategory

```python
df2.loc[df2['Age'] >= 16, ['AgeCategory']] = 'Adult'
df2.loc[df2['Age'] < 16, ['AgeCategory']] = 'Child'
```

For passengers whose age we did not know, we will use the other clues discussed to derive AgeCategory.

```python
# If Mr./Mrs. honorific is used in the name, AgeCategory = 'Adult'
df2.loc[df2['Name'].str.contains('Mrs\.|Mr\.'), ['AgeCategory']] = 'Adult'

# If Master. honorific is used in the name, AgeCategory = 'Child'
df2.loc[df2['Name'].str.contains('Master\.'), ['AgeCategory']] = 'Child'

# For each passenger, calculate the number of travelling companions
df2['Companions'] = df.SibSp + df.Parch

# If Companions = 0, AgeCategory = 'Adult'
df2.loc[df2['Companions'] == 0, ['AgeCategory']] = 'Adult'
```

Let's take a look at how many passengers still do not have a value for AgeCategory

```python
df2[df2['AgeCategory'].isnull()]
```

We now have an age category for all but 14 passengers, which is very good considering we are missing an age value for 177 passengers. All of these remaining 14 passengers use the honorific 'Miss.', which is not a strong indicator of relative age. It only implies a female's marital status of being single, which could be a female of any age. Since we have no data left to indicate the age category for these passengers, and it is an insignificant number of records relative to the total number, we can drop these passengers from our age analysis.

```python
df2.drop(df2[df2.AgeCategory.isnull()].index, axis=0, inplace=True)
```

Now we can look for survivors based on their age category

```python
df2.groupby(by=['Survived','AgeCategory']).AgeCategory.count()
```

From the above we can see that a greater percentage of children (59%) survived than adults (36%).
